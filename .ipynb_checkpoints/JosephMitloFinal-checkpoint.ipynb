# Importing the Data

```python
#import the data
import pandas as pd
import geopandas as gp
%matplotlib inline
data = pd.read_csv("courts-and-rinks.csv")
pd.set_option('display.max_rows', 10)
```

# Let's get data!

```python
#make a graph showing the top 10 places with the most courts
graph = data['neighborhood'].value_counts()[:10]
graph.plot.barh()

#print out the counts of the places wotht he most courts
data['neighborhood'].value_counts()
```

# Making a graph

```python
data['field_count'] = 1
type_fields = data.groupby("neighborhood").sum()['field_count']
neighborhoods = gp.read_file("Neighborhoods/Neighborhoods_.shp")
```

```python
fields_map = neighborhoods.merge(type_fields,how='left', left_on='hood', right_on='neighborhood')
fields_map.plot(column='field_count',
                cmap='YlGn',
                edgecolor="white",
                legend=True,
                legend_kwds={'label': "Number of Fields"},
                figsize=(10,5),
                missing_kwds={"color": "lightgrey"}
               )
```
